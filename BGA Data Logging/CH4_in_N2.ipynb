```python
# Import libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Path to file that will be loaded
file_path = 'P517 Data Logger Spit Out Trial 3.txt'

# Load txt file as csv into a dataframe
df = pd.read_csv(file_path)

# Convert time column to datetime objects for simpler plotting
df['Time'] = pd.to_datetime(df['Time'], format='%Y-%m-%d %H:%M:%S')

print(df)
```

```python
# Generate labelled plot of the data
plt.plot(df['Time'], df['Current(%)'])
plt.xlabel('Day/Time')
plt.ylabel('% of CH4 in N2')
plt.title('Percentage of methane in nitrogen')

# Add red markers to data that is below expected experimental threshold
lower_threshold = df['Current(%)'] < 5
plt.scatter(df['Time'][lower_threshold], df['Current(%)'][lower_threshold], color='red')

# Add red markers to data that is above expected experimental threshold
upper_threshold = df['Current(%)'] > 22
plt.scatter(df['Time'][upper_threshold], df['Current(%)'][upper_threshold], color='red')

```
